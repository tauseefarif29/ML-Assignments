### Problem Statement:
#### In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.
#### The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

Visualizing the image before compressing

```python
Racoon = scipy.misc.face(gray=True)
plt.figure(figsize=(10,5))
plt.imshow(Racoon,plt.cm.gray)
plt.show()
```

Getting rows and columns of the image

```python
rows = Racoon.shape[0]
columns = Racoon.shape[1]
image = Racoon.reshape(rows*columns,1)
```

#### Compressing the image using KMeans to fit the image

```python
Kmeans = KMeans(n_clusters=5)
Kmeans.fit(image)
```

Getting the cluster and the labels for the image

```python
clusters = np.asarray(Kmeans.cluster_centers_)
labels = np.asarray(Kmeans.labels_)
labels = labels.reshape(rows,columns)
```

Save the compressed image

```python
plt.imsave('compressed_racoon.png',labels)
```

Visualize the compressed image

```python
image = plt.imread('compressed_racoon.png')
plt.figure(figsize=(10,5))
plt.imshow(image)
plt.show()
```
